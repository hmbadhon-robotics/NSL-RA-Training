# Normalizer

```python
df = pd.DataFrame({
    'x1': np.random.randint(-100, 100, 1000).astype(float),
    'y1': np.random.randint(-80, 80, 1000).astype(float),
    'z1': np.random.randint(-150, 150, 1000).astype(float),
})
```

### Before

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(df.x1, df.y1, df.z1)
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
data_tf = normalizer.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x1','y1','z1'])
```

### After

```python
ax = plt.axes(projection='3d')
ax.scatter3D(df.x1, df.y1, df.z1)
```

### Using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df = pd.read_csv("Fish.csv")
display(df)
```

```python
le=LabelEncoder()
df['Encoded_Species']=le.fit_transform(df.Species)
```

```python
display(df)
```

### OneHotEncoder

```python
# Encoding categorical features 
enc = OneHotEncoder()
categ = df[['Species']]
encoded_categ = pd.DataFrame(enc.fit_transform(categ).toarray())


#Likning the encoed_cateh with the df
df = pd.concat([df, encoded_categ], axis = 1)
display(df)

#droping columns
#df.drop(['C', 'D'], axis = 1)
cols=['Species','Encoded_Species']
df.drop(cols, axis = 1, inplace = True)
display(df)
```

# Polynomial
An input sample is two dimensional and of the form \[a, b\], the degree-2 polynomial features are \[1, a, b, a^2, ab, b^2\].

```python
df = pd.DataFrame({'A':[1,2,3,4,5], 'B':[2,3,4,5,6]})
display(df)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol = PolynomialFeatures(degree=2)
poll = pol.fit_transform(df)
display(pd.DataFrame(poll))
```

# Custom Transformer

```python

from sklearn.preprocessing import FunctionTransformer
```

```python
df=pd.read_csv("Fish.csv")
display(df)
```

```python
def mapping(x):
    x['Weight'] = x['Weight']/x["Weight"].mean()
    x['Length4'] = x['Length1']+x['Length2']+x['Length3']
    return x
```

```python
customtransformer = FunctionTransformer(mapping, validate=False)
```

# Count Vectorizer

```python
corpus = [
     'This is the first document awesome food.',
     'This is the second second document.',
     'And the third one the is mission impossible.',
     'Is this the first document?',
]
df = pd.DataFrame({'Text':corpus})
display(df)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit_transform(df.Text).toarray()
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(vocabulary=['mission','food','second'])
cv.fit_transform(df.Text).toarray()
```

```python
cv = CountVectorizer(ngram_range=[1,2])
cv.fit_transform(df.Text).toarray()
```

```python
cv.vocabulary_
```
