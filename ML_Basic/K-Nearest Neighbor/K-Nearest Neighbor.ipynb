```python
#Select Features
X = df.drop(columns =['Outcome'], axis = 1)

#Select Target 
y = df['Outcome'].values
```

# DATASET SPLITING

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)

# Finding Best Parameter(k value as neighbours)

```python
# try K=1 through K=25 and record testing accuracy
k_range = range(1, 50)

# We can create Python dictionary using [] or dict()
scores = []
score_neighbour=[]

# We use a loop through the range 1 to 26
# We append the scores in the dictionary
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))
    score_neighbour.append([metrics.accuracy_score(y_test, y_pred),k])
    

#print(scores)
#print(score_neighbour)
score_neighbour.sort(key=lambda x: x[0])
#print(score_neighbour[-1])
```

```python
plt.plot(k_range, scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

# Best MODEL 

```python
knn = KNeighborsClassifier(n_neighbors=score_neighbour[-1][1])
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

# MODEL EVALUATION

```python
print(metrics.classification_report(y_test, y_pred))
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score Naive Bayes: {0}'.format(metrics.accuracy_score(y_test, y_pred))
plt.title(all_sample_title, size = 15);
```
