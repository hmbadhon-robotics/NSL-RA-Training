# Dataset Loading
This is a dataset about fish and it's some features like length,height.
I have tried to predict the name of the fish from multiple attributes with multiple linear regression.

```python
df = pd.read_csv("Fish.csv")
display(df)
```

```python
df["Species"].replace({"Perch": 1, "Bream": 2,"Roach": 3,"Smelt": 4,"Parkki":5,"Whitefish":6,"Pike":7}, inplace=True)
print(df)
```

```python
X = df[[ 'Weight', 'Length1', 'Length2', 'Length3', 'Height',
       'Width']].values  
#.values is converting dataframe into array
print(X)
y = df[['Species']].values
print(y)
```

```python
print(y.shape)
```

# DATASET SPLITING

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# MODEL 

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(X_test[0:1])
sample_predict=model.predict(X_test[0:1])
print(sample_predict)
```

# MODEL EVALUATION

```python
#Comparison
print('- y_pred : ')
print(y_pred)
print('- y_test : ')
print(y_test)
```

```python
print(y_pred.sum().mean())
print(y_test.sum().mean())
print("So called Accuracy ", y_pred.sum().mean()/y_test.sum().mean())
```
